```python
# Các bước chính:
#     1.Đọc dữ liệu và chia train test(lấy toàn bộ dữ liệu có)
#     2.Tìm Bag of Word và tìm tần suất xuất hiện của các từ trong từng văn bản
#     3.Tìm các xác suất cần thiết( có làm trơn xác suất )
#     4. Thực hiện predict và hiển thị kết quả
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv("D:\\Project\\AI\\EmailsClean.csv")
df['email'] = df['emailClean'].apply(lambda x: x.split(" "))
df.columns = ['index','spam','emailClean','email']
df.drop(columns = ['index','emailClean'],inplace = True)
spam = df[df['spam'] == 1]
notSpam = df[df['spam'] == 0]
l_train_spam = int(len(spam)*0.7)
l_train_notSpam = int(len(notSpam)*0.7)
#chia train test với train 70% data
X_train = spam[0:l_train_spam].append(notSpam[0:l_train_notSpam],ignore_index = True)
X_test = spam[l_train_spam:len(spam)].append(notSpam[l_train_notSpam:len(notSpam)],ignore_index = True)
```

# Create Bag Of 

```python
# create Bag of Word full
BoW = set()
for doc in X_train.email:
    for word in doc:
        BoW.add(word)
len(BoW)
```

# Tìm tần suất xuất hiện các từ

```python
bayes_matrix = np.zeros((X_train.shape[0],len(BoW)))
frequency = pd.DataFrame(data = bayes_matrix)
frequency.columns = list(BoW)
```

```python
#tạo bảng tần suất các từ trong câu 
frequency1 = frequency.loc[0:l_train_spam-1] 
frequency0 = frequency.loc[l_train_spam:len(X_train)]
frequency1
```

```python
frequency0
```

```python
# tìm tần suất
for i in range(0,l_train_spam):
    for word in X_train.email[i]:
        if word in BoW:
            frequency1[word][i] = frequency1[word][i] + 1.0
for i in range(l_train_spam,len(X_train)):
    for word in X_train.email[i]:
        if word in BoW:
            frequency0[word][i] = frequency0[word][i] + 1.0
```

# Tổng số lần xuất hiện của một từ trong các văn bản spam

```python
total1 = pd.DataFrame(data = frequency1.apply(np.sum,axis = 0))
total0 = pd.DataFrame(data = frequency0.apply(np.sum,axis = 0))
```

```python
total1
```

```python
# tổng số từ trong các văn bản spam
N1= int(total1.apply(np.sum,axis = 0)) 
N0= int(total0.apply(np.sum,axis = 0))
```

# Measure P(c),P(x|c)

```python
# bảng tổng hợp các xác suất 1 từ thuộc nhãn 0 hoặc 1
Probability_matrix = np.zeros((2,len(BoW)))
Probability = pd.DataFrame(data = Probability_matrix)
Probability.columns = list(BoW)
Probability 
```

```python
P_spam = float((l_train_spam)/(l_train_spam+l_train_notSpam))
P_nonSpam = float((l_train_notSpam)/(l_train_spam+l_train_notSpam))
P_spam #P(c)
```

```python
#làm trơn xác suất
#P(xi|c) = (total+1)/(N+len(BoW))
#total là số lần xuất hiện của từ (nhãn 0|1)
#N tổng số từ của tất cả các văn bản(nhãn 0 hoặc 1)
#len(BoW): số từ trong bag of word
for word in BoW:
    Probability[word][1] = ((float(total1.loc[word,:])+1.0)/(N1+len(BoW)))
    Probability[word][0] = ((float(total0.loc[word,:])+1.0)/(N0+len(BoW)))
Probability
```

# Predict

```python
import math
def predict(email):
    P_1 = 0.0
    P_0 = 0.0
    for word in email:
        if word in BoW:
            P_0 = P_0+math.log10(Probability[word][0])
            P_1 = P_1+math.log10(Probability[word][1])
        else:
            P_0+=1.0/(N0+len(BoW))
            P_1+=1.0/(N1+len(BoW))
    P_0 = math.log10(P_nonSpam) + P_0
    P_1 = math.log10(P_spam) + P_1
    if P_1 > P_0:
        return 1
    else:
        return 0
```

# Result

```python
from sklearn.metrics import accuracy_score
pred = pd.DataFrame(data = [predict(d) for d in X_test.email])
accuracy_score(X_test.spam, pred)
#F1 score
```

```python
from sklearn.metrics import confusion_matrix
cnf_matrix = confusion_matrix(pred,X_test.spam)
print('Confusion matrix:')
print(cnf_matrix)
```
