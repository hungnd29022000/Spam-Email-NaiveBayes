# Clean Data

```python
import pandas as pd
import numpy as np
```

```python
#load dataset
df = pd.read_csv("D:\\Project\\AI\\emails.csv")
df.head()
```

```python
import re
from gensim.parsing.preprocessing import strip_non_alphanum, strip_multiple_whitespaces,preprocess_string, split_alphanum, strip_short, strip_numeric
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
sno = nltk.stem.SnowballStemmer('english')
#lấy các stopword để loại bỏ
```

```python
def clean_mail(document):
    document = document.lstrip('Subject: ')#xóa từ Subject: ở đầu mỗi văn bản
    document = re.sub(r"http\S+", "", document)#xóa link
    document = strip_non_alphanum(document).lower().strip()#loại bỏ các kí tự đặc biệt
    document = split_alphanum(document)#tách bỏ các từ ghép với số
    document = strip_short(document, minsize=2)#loại bỏ các từ đơn 
    document = strip_numeric(document)#loại bỏ các số 
    #lọc bỏ stopword
    document = [word for word in document.split() if word.lower() not in stopwords.words('english')]
        document[i] = sno.stem(document[i])
    document = " ".join(document)
    return document
```

```python
#loại bỏ các bản ghi trùng nhau trong dữ liệu
df.drop_duplicates(inplace = True)
df
```

```python
#thực hiện clean từng data
df['emailClean'] = df['text'].apply(clean_mail)
```
